```python
# NOTE: FOR FUTURE USE, THE 'PLAYLIST' COLUMN MUST BE ADDED, AND THE SHAPE MUST BE CHECKED.
allplaylists = []
i = 0
for playlist in testpath.iterdir():
    temp = pd.read_csv(playlist, encoding="ISO-8859–1")
    temp['Playlist'] = str(i+1)
    temp = temp.iloc[:, 1:]
    if temp.shape[0] > 0:
        allplaylists.append(temp)
    i+=1
```

# Processing 2 - Utilize Linear Algebra to Assess Robustness of Multiclass Classification

```python
def dataprocess1(dataframe):
    y = dataframe['Playlist']
    X = dataframe.drop(columns=['type', 'id', 'uri', 'track_href', 'analysis_url', 'time_signature', 'key', 'Playlist', 'mode'])
    X = (X - X.mean()) / X.std()
    return X, y
```

# 

```python
X, y = dataprocess1(playlists)
X
```

```python
u, s, vt = np.linalg.svd(X)
```

```python
plt.stem(s**2/sum(s**2))
plt.title('% Of Variance Attributable to Each Feature')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Relatively promising - we have one strong feature and a handful of moderately effective features
lr = LogisticRegression(C = 0.5, multi_class="multinomial", max_iter=1000)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3)
```

```python
lr.fit(X_train, Y_train)
```

```python
X_test.iloc[0,:]
```

```python
lr.score(X_test, Y_test)
```
